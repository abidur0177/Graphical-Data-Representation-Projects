```python
#importing all the required libraries

import pandas as pd #the pandas library is used to manipulate dataframe. imported as pd
import seaborn as sns #seaborn is a plotting library. imported as sns
import numpy as np #numpy is for numerical data analysis. imported as np
import matplotlib.pyplot as plt #matplotlib is also plotting library. pyplot is a submodule. imported as plt
%matplotlib inline
```

```python
#read the excel file
df = pd.read_excel('heatmap.xlsx') 
df.head() #reading the first 5 rows. df.head(10) means showing the first 10 rows
```

```python
#Check the shape
df.shape
#there are 1922 rows and 452 columns in this dataframe
```

```python
#check the info
df.info()
#this provides information of every column
```

```python
#check the null values
df.isnull().any().sum() 
#no null values are there. the above function provides the summation of null cells in the dataframe
```

```python
#rename the first column
df = df.rename(columns={'Position [mm]/Time [s]':'0'})
#this rename function is used to rename column names. Our first column is Position [mm]/Time [s], so we replace the
#name with 0
```

```python
#set the first column as index and drop the index of the dataframe
df.set_index('0', drop=True, inplace=True)
df
#now set_index function is used to set any column as index. since we want our first column in the y axis so we need
#to set the 0 named column as our index and drop=True means drop/remove the original index of the dataframe and
#inplace=True means set it as our final dataframe
```

### Plotting the heatmap

```python
fig, ax = plt.subplots(figsize=(18,8)) #creating a figure and an axis (fig, ax) using plt. set the figure size of
#(18,8). width=18, height=8

#plotting the heatmap. the first argument df.iloc[:,:122] is the dataframe. iloc is used to take subset of the original
#data. the first argument is the number of rows we want. since you want every rows I put : 
#and you want 18150 in the xaxis so :122 means selecting the columns upto 122.
#The cmap allows your imply your preferred color theme onto the canvas. 
#fmt means fonts formatting.
#cbar_kws is used to modify the Cbar of the heatmap. location=bottom means the position of the cbar and label
#means the label of the cbar
sns.heatmap(df.iloc[:, :122], cmap='Greys_r', fmt=".2f", cbar_kws = dict(use_gridspec=False,
                                                           location="bottom", label='Transmission, %'))
    
#x axis ticks edit
ax.xaxis.tick_bottom() #position the x axis at the bottom for better view
plt.xticks(rotation=75) #rotation of the xaxis tick labels. we use 75 degree rotation



#roundind the y axis. remove the decimal of yticks and plot the rounded figure
labels = [item for item in ax.get_yticklabels()] #ax.get_ytickslabels() return the yaxis labels in text format
                                                 #we get all the label in text format and input into a list named labels

labels_str = [str(i) for i in labels] #converting the elements of labels from text format to string format

label_only_value_first = labels_str[0][-9:-6] #get the first element of labels_str list and extract the rounded value
                                              #from the string. We did this explicitely for only the first element
                                              #because the first string value length is different than others

labels_only_value=[] #define an empty list
labels_only_value.append(label_only_value_first)#insert the first rounded value into the list named labels_only_value.

for i in labels_str[1:]: #run a for loop in the labels_str excluding the first element as it is already extracted
    val_str = i[-10:-7] #get the rounded value and store it in the val_str variable
    labels_only_value.append(val_str) #insert into the list named labels_only_value
    
labels_only_value[13] = '115' #the 13th indexed value is empty in the labels_only_value list. so change it to 115
labels_only_value[26] = '126' ##the 26th indexed value is empty in the labels_only_value list. so change it to 126
labels_int = [int(i) for i in labels_only_value] #loop over the labels_only_value and convert each element as integer and
                                                 #store these in the list named labels_int
    
ax.set_yticklabels([label for label in labels_int]) #set the rounded labels and the yaxis tick labels


#Axis Labels
plt.xlabel('Time', fontsize=25) #label the xaxis as Time
plt.ylabel('Position',fontsize=25) #label the yaxis as Position


plt.show() #showing the plot
```
