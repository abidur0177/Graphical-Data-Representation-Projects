### 2. Create a new Column (Named:"CDM_Name") based on the on the following criteria: 
        if a Row in "location" is ("96%H2SO4 AQRM 01 On Line") then new name is ("CDM_1")
        if a Row in "location" is ("96%H2SO4 AQRM 02 On Line") then new name is ("CDM_2")
        if a Row in "location" is ("96%H2SO4 AQRM 03 On Line") then new name is ("CDM_3")
        if a Row in "location" is ("96%H2SO4 AQRM 04 On Line") then new name is ("CDM_4")

```python
#creating the CDM_NAME column
def CDM(x):
    if x == '96%H2SO4 AQRM 01 On Line' :
        return 'CDM1'
    elif x == '96%H2SO4 AQRM 02 On Line':
        return 'CDM2'
    elif x == '96%H2SO4 AQRM 03 On Line':
        return 'CDM3'
    elif x == '96%H2SO4 AQRM 04 On Line':
        return 'CDM4'
    return 'other'

df['CDM_Name'] = df['location'].apply(lambda x: CDM(x))
```

###  4. Reshape the df table ("df_Totes_CDM_Day" ) so it will looks like below:

```python
from collections import OrderedDict 
z = df_Totes_CDM_Day.groupby(['DMY'])['CDM_Name'].value_counts().to_dict()
sort = OrderedDict(sorted(z.items()))
sort = dict(sort)
sort[('10/02/2021', 'CDM2')] = 0
sort[('12/02/2021', 'CDM2')] = 0
sort = OrderedDict(sorted(sort.items()))
sort = dict(sort)
sort
```

```python
keys = list(sort.keys())
values = list(sort.values())
CDM_2 =[]
CDM_3 = []
CDM_4 = []
for i in range(0,len(values)-2,3):
    CDM_2.append(values[i])
    CDM_3.append(values[i+1])
    CDM_4.append(values[i+2])
print(len(values))
```

```python
len(CDM_2), len(CDM_3), len(CDM_4)
```

```python
keys_non_dup = []
for i in range(len(keys)):
    if str(keys[i][0]) in keys_non_dup:
        continue
    else:
        keys_non_dup.append(keys[i][0])
keys_non_dup
keys_non_dup
```

```python
df_Totes_CDM_Day = pd.DataFrame({'DMY':keys_non_dup, 'CDM_2':CDM_2, 'CDM_3':CDM_3,'CDM_4':CDM_4}, index=None)
df_Totes_CDM_Day
```
