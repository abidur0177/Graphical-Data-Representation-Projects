### Final Testing Part

### Limiting the x axis

```python
# get a date range from two specific date  to limit the x axis
datestart = datetime.datetime(2020,3,10)
dateend = datetime.datetime(2020,7,11)
delta = dateend - datestart #timedelta
print(delta)
date_range = [] #take the date range between two individual date
for i in range(0,delta.days+1, 1):
    date_range.append((datestart + timedelta(days=i)))
date_range
```

```python
#get the dataframe within the date range
subset = df.loc[df['DMY'].isin(date_range)]
subset
```

### X ticks

```python
#now get dates for xticks whithin the date_range
datestart_xticks = subset['DMY'].iloc[0]
dateend_xticks = subset['DMY'].iloc[-1]
delta_xticks = dateend_xticks - datestart_xticks
print(delta_xticks.days)
xticks_all_date = []
for j in range(0,delta_xticks.days+1, 4):
    xticks_all_date.append((datestart_xticks + timedelta(days=j)))
#xticks_all_date
```

```python
#now check where the each date in xticks_all_date belogs to the subset
final_xticks_timestamp = []
subset_for_xticks = subset.loc[subset['DMY'].isin(xticks_all_date)]
for k in subset_for_xticks['DMY']:
    if k in final_xticks_timestamp:
        continue
    else:
        final_xticks_timestamp.append(k)
final_xticks = []
for i in final_xticks_timestamp:
    final_xticks.append(i.date())
final_xticks
```
